```r
library(RSQLite)
```

```r
con = dbConnect(drv=RSQLite::SQLite(), dbname="database.sqlite")
```

```r
alltables = dbListTables(con)
```

```r
alltables
```

```r
alltables <- alltables[alltables != "sqlite_sequence"]

lDataFrames <- vector("list", length=length(alltables))
```

```r
for (i in seq(along=alltables)) {
  lDataFrames[[i]] <- dbGetQuery(conn=con, statement=paste("SELECT * FROM '", alltables[[i]], "'", sep=""))
}
```
